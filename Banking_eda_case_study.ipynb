```
bins = [0, 100000, 300000, float('inf')]
labels = ['low', 'medium', 'high']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```
df['Income Band'].value_counts().plot(kind='bar')
```

```
from pandas.core.arrays import categorical
#Examine the distribution of unique categories in categorical coloumns
categorical_col = df[[
    "BRId",
    "GenderId",
    "IAId",
    "Amount of Credit Cards",
    "Nationality",
    "Occupation",
    "Fee Structure",
    "Loyalty Classification",
    "Properties Owned",
    "Risk Weighting",
    "Income Band"
]]

for col in categorical_col.columns:
    print(f"\n{'-'*40}")
    print(f"Value Count for '{col}'")
    print(categorical_col[col].value_counts(dropna=False))
```

UNIVARIATE ANALYSIS
  Univariate = Uni = One

You analyze a single variable at a time to understand:


```
for col in categorical_col.columns:
    plt.figure(figsize=(8, 4))
    categorical_col[col].value_counts().plot(kind="bar")

    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```

BIVARIANTE ANALYSIS

Bivariate = Bi = Two

You analyze two variables together to understand:


```


nationality_count = df["Nationality"].value_counts()

plt.figure(figsize=(10,5))
nationality_count.plot(kind="bar")
plt.title("Customer Count by Nationality")
plt.xlabel("Nationality")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()



```

```
#histo plot for value count
occupation_counts = df["Occupation"].value_counts()

plt.figure(figsize=(8,5))
plt.hist(occupation_counts, bins=10)
plt.title("Histogram of Customer Count per Occupation")
plt.xlabel("Number of Customers per Occupation")
plt.ylabel("Number of Occupations")
plt.tight_layout()
plt.show()
```

```
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns
numerical_cols
```

```
df[numerical_cols].describe()
```

```
df[numerical_cols].isna().sum()
```

```
import matplotlib.pyplot as plt

for col in numerical_cols:
    plt.figure(figsize=(7,4))
    plt.hist(df[col].dropna(), bins=15)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Number of Customers")
    plt.tight_layout()
    plt.show()

```

```
correlation_matrix = df[numerical_cols].corr()
correlation_matrix
```

```
plt.figure(figsize=(10,8))
plt.imshow(correlation_matrix)
plt.colorbar()
plt.xticks(range(len(numerical_cols)), numerical_cols, rotation=90)
plt.yticks(range(len(numerical_cols)), numerical_cols)
plt.title("Correlation Matrix of Numerical Variables")
plt.tight_layout()
plt.show()
```

HEATMAPS


```


numerical_cols = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10,8))
sns.heatmap(
    numerical_cols.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Correlation Heatmap of Numerical Variables")
plt.tight_layout()
plt.show()
```
